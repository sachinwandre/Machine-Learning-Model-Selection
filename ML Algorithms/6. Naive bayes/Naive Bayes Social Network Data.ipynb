```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report,confusion_matrix,recall_score,accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
data  = pd.read_csv('D:/14_My Practice/Datasets/Social_Network_Ads.csv')
data
```

```python
X = data.drop(['User ID','Gender','Purchased'],axis=1)
y = data.Purchased
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=30)
```

```python
standscale = StandardScaler()
standscale.fit(X_train)
X_train_scaled =  standscale.transform(X_train)
X_test_scaled = standscale.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
```

```python
X_test_scaled
```

```python
X_train_scaled
```

```python
standscale.mean_
```

```python
X_test_scaled
```

```python
X_train_scaled
```

```python
nbc = GaussianNB()
nbc.fit(X_train_scaled,y_train)
```

```python
y_pred = nbc.predict(X_test_scaled)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(recall_score(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
